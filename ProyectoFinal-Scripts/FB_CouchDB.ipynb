```python
from facebook_scraper import get_posts
import couchdb
import json
import time
import pandas as pd
```

```python
couch=couchdb.Server('http://poleth:poleth30032001@127.0.0.1:5984')
```

```python
nombredb='datos'
db=couch[nombredb]
```

```python
i=1
for post in get_posts('microsoft', pages=5, extra_info=True):
    print(i)
    i=i+1
    time.sleep(5)
    
    id=post['post_id']
    doc={}
     
    doc['id']=id
    
    mydate=post['time']
    
    try:
        doc['texto']=post['text']
        doc['date']=mydate.timestamp()
        doc['likes']=post['likes']
        doc['comments']=post['comments']
        doc['shares']=post['shares']
        try:
            doc['reactions']=post['reactions']
        except:
            doc['reactions']={}

        doc['post_url']=post['post_url']
        db.save(doc)
        frame=pd.DataFrame(doc, index=[1])
        frame.to_csv(str(id)+'.csv')

    
        print("guardado exitosamente")

    except Exception as e:    
        print("no se pudo grabar:" + str(e))
```
